Сравним полученные значения со стационарными формулами:

```python
import math as m
# Вариант №1: 
# Уравнение притока к горизонтальной сважине, расположенной 
# в центре однородного изотропного пласта с круговым контуром питания, приведено 
# в работе (Борисов и др., 1964)
pi = 3.14
Rк = 750 # м, радиус контура питания
dP = 7*10**6 # Па, перепад давления
k = 100/10**15 # м2, проницаемость (в числовой модели изотропный пласт)
h = 10 # м, толщина пласта
L = 60 # м, длина горизонтального участка в пласте
mu = 0.0005 # Па*с, вязкость
rc = 0.078 # м, радиус скважины
Q1 =2*pi*k*h/mu * dP/(m.log1p(4*Rк/L)+h/L*m.log1p(h/2/pi/rc))*86400
print('Приток по первому варианту: ' + str(Q1) + ' м3/сут')

# Вариант №2:
# В работе (Renard & Dupuy, 1991) показано формула для дебита горизонтальной скважины 
# в  элептическом и прямоугольном контурах питания:

#betta = (kh/kv)^0.5 
betta = 1 # коэффициент анизотропии проницаемости;
r_ = rc*(1+betta)/(2*betta)
a = 1500 # м, сторона модели
X = 2*a/L # для элептического контура питания
Q2 = 2*pi*k*h/mu * dP/(1/m.cos(X)+betta*h/L*m.log1p(h/(2*r_)))*86400
print('Приток по второму варианту: ' + str(Q2) + ' м3/сут')
```
